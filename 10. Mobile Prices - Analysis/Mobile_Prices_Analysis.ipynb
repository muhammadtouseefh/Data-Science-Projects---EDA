# **Context**
Bob has started his own mobile company. He wants to give tough fight to big companies like Apple,Samsung etc.

He does not know how to estimate price of mobiles his company creates. In this competitive mobile phone market you cannot simply assume things. To solve this problem he collects sales data of mobile phones of various companies.

Bob wants to find out some relation between features of a mobile phone(eg:- RAM,Internal Memory etc) and its selling price.

For more info on the dataset and feature meanings, check this [Kaggle Dataset](https://www.kaggle.com/datasets/iabhishekofficial/mobile-price-classification?select=train.csv) out.

```python
# Load needed libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from urllib.request import urlretrieve 
print("Imported Needed Libraries 😅")

# Set Settings
pd.set_option('display.max_rows', 2000) 
pd.set_option('display.max_columns', 21)

# Get the data
URL = "https://raw.githubusercontent.com/muhammadanas0716/Data-Science-Projects---EDA/main/10.%20Mobile%20Prices%20-%20Analysis/Mobile%20Prices.csv"
mobile_df = pd.read_csv(URL)

print("Data Imported Successfully 🚀")
```

```python
# Change Column Names and Data

# Bluetooth Column
mobile_df.rename(columns={'blue': 'bluetooth'},
          inplace=True, errors='raise')
mobile_df["bluetooth"].replace({0: "No", 1: "Yes"}, inplace=True)

# Dual Sim
mobile_df["dual_sim"].replace({0: "No", 1: "Yes"}, inplace=True)

# 4G
mobile_df["four_g"].replace({0: "No", 1: "Yes"}, inplace=True)

# 3G
mobile_df["three_g"].replace({0: "No", 1: "Yes"}, inplace=True)

# TouchScreen
mobile_df["touch_screen"].replace({0: "No", 1: "Yes"}, inplace=True)

# Wifi
mobile_df["wifi"].replace({0: "No", 1: "Yes"}, inplace=True)

# Price Range
mobile_df["price_range"].replace({0: "Low Cost", 1: "Medium Cost", 2 : "High Cost", 3 : "Very Hight Cost"}, inplace=True)

print("Columns and Data Updated Successfully 🤯")
```

```python
# Get the highest correlated columns to PRICE RANGE
corrmat = mobile_df.corr()
figure, ax = plt.subplots(figsize=(15, 12), dpi=100)
sns.heatmap(corrmat, vmax=.8, square=True, annot=True)
plt.xticks(rotation=360)
plt.show()
```

## Analysis ⬆️

We can obeserve that the Camera functunalites have a huge impact on the Price.

**Those columns (CAMERA RELEATED) are below**
1. fc ➡️ Front Camera
2. px_height ➡️ Pixel Resolution Height
3. px_width ➡️Pixel Resolution Width

**And Also the Phone dimensions**
1. sc_h ➡️Screen Height of mobile in cm
2. sc_w ➡️Screen Width of mobile in cm

***For more info on the dataset and feature meanings, check this [Kaggle Dataset](https://www.kaggle.com/datasets/iabhishekofficial/mobile-price-classification?select=train.csv) out.***

```python
# Let's get some count plots - FUNCTION CREATION

def count_plot(feature_name, hue_feature="price_range", title=None):
  plt.style.use("fivethirtyeight")
  figure, ax = plt.subplots(figsize=(10, 6), dpi=100)
  plt.title(title)
  sns.countplot(data=mobile_df, x=feature_name, hue=hue_feature)
  plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
  plt.show()
```

## Note ⬆️
 
I did not want to repeat the same thing over and over again, otherwise this notebook would get very legnthy for no optimal reason. Just use the function **count_plot** to show a count plot for a specific feature. This will save time and code as well as runtime. 

Creating fucntions like this is a very good paratice as repeating your code is considered to be a menace.

```python
# Count Plot FOR Having 3G or NOT
count_plot("three_g", "price_range", "Having 3G or NOT VS Sale Price") # Function Definition
```

## Analysis ⬆️

We observe that most phones do come with 3G support. Those phones that DON'T come with 4G support are relatively in lower amount than those which do COME with 3G support. Meaning most phones WILL and DO have 3G support

```python
# Count Plot FOR Having 4G or NOT
count_plot("four_g", "price_range", "Having 4G or NOT VS Sale Price") # Function Definition
```

## Analysis ⬆️

We observe that half of the phones do come with 4G support, the rest don't. Nonetheless, price range does not get affected too much.

```python
# Count Plot FOR Supporting Bluetooth or NOT
count_plot("bluetooth", "price_range", "Supporting Bluetooth or NOT") # Function Definition
```

## Analysis ⬆️

We observe that with our without Bluetooth, they don't differ too much on different price ranges.  

Only **Very High Cost** has some variation, as in NO bluetooth - they are less phones rather than WITH BLUETOOTH, which have the highest count there(RIGHT-HAND-SIDE).


```python
# Relation of Pixel Resolution Height and Pixel Resolution Width with Price Ranges 
plt.figure(figsize=(20,10))
sns.scatterplot(x = "px_height", y = "px_width", data = mobile_df, hue = "price_range").set(
                     title = "Relation of Pixel Resolution Height with Price Ranges",
                     xlabel = "Pixel Resolution Height",
                     ylabel = "Pixel Resolution Width")

plt.show()
```

```python
# Relation of Screen Height and Screen Width with Price Ranges 
plt.figure(figsize=(20,10))
sns.scatterplot(x = "sc_h", y = "sc_w", data = mobile_df, hue = "price_range").set(
                     title = "Relation of Screen Height and Screen Width with Price Ranges ",
                     xlabel = "Screen Height",
                     ylabel = "Screen Width")

plt.show()
```
